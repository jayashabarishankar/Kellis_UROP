```python
from Bio import SeqIO
import numpy as np
```

```python
def prob_matrix(filename):
#filename= "/Users/jayas/Kellis_UROP/test.fa" directory for my test file
    sequences= [i for i in SeqIO.parse(filename, "fasta")] #list of SeqIO objects
    assert len(sequences)==2 #2 sequences will be compared
    list_of_sequences=[]
    list_of_indices=[]
    for index in range(0,len(sequences[0])):
        if sequences[0][index]!="N" and sequences[0][index]!="-":
            list_of_indices.append(index) #creates a list of indices in the first sequence that do not include N or - spots
            
    for index in list_of_indices:
        if sequences[1][index]=="N" or sequences[1][index]=="-":
            list_of_indices.remove(index) #ensures that the list of indices do not include N or - in the second sequence either
    seq1=[]
    seq2=[]
    for ind in list_of_indices:
        seq1.append(sequences[0][ind])
        seq2.append(sequences[1][ind]) #new lists made with all unique nucleotides
    
    list_of_sequences.append(seq1)
    list_of_sequences.append(seq2)

    #ACGT order will be followed, sequence 1 is the one on top/ corresponding to columns
    prob_matrix = np.zeros(shape=(4, 4)) 

    count_ag=0
    count_ac=0 #these are counts from the first nucleotide changing to the second nucleotide in the second sequence
    count_at=0


    count_ga=0
    count_gc=0
    count_gt=0

    count_cg=0
    count_ca=0
    count_ct=0

    count_ta=0
    count_tc=0
    count_tg=0
    length_1=len(seq1)
    length_2=len(seq2)
    if length_2>length_1:
        list_of_sequences.append(seq_1)
        list_of_sequences.remove(seq_1) # to ensure that the first sequence in list of sequences is the shorter one
    for nuc_index in range(0,len(list_of_sequences[0])):
        if list_of_sequences[0][nuc_index]!=list_of_sequences[1][nuc_index]: #loop through every nucleotide in the shorter sequence
            if list_of_sequences[0][nuc_index]=="A":
                if list_of_sequences[1][nuc_index]=="G":
                        count_ag+=1
                if list_of_sequences[1][nuc_index]=="C":
                        count_ac+=1
                if list_of_sequences[1][nuc_index]=="T":
                        count_at+=1
            if list_of_sequences[0][nuc_index]=="G":
                if list_of_sequences[1][nuc_index]=="A":
                        count_ga+=1
                if list_of_sequences[1][nuc_index]=="C":
                        count_gc+=1
                if list_of_sequences[1][nuc_index]=="T":
                        count_gt+=1
            if list_of_sequences[0][nuc_index]=="C":
                if list_of_sequences[1][nuc_index]=="G":
                        count_cg+=1
                if list_of_sequences[1][nuc_index]=="A":
                        count_ca+=1
                if list_of_sequences[1][nuc_index]=="T":
                        count_ct+=1
            if list_of_sequences[0][nuc_index]=="T":
                if list_of_sequences[1][nuc_index]=="A":
                        count_ta+=1
                if list_of_sequences[1][nuc_index]=="C":
                        count_tc+=1
                if list_of_sequences[1][nuc_index]=="G":
                        count_tg+=1
                    
    prob_matrix[0,1]=count_ca/len(list_of_sequences[0])
    prob_matrix[0,2]=count_ga/len(list_of_sequences[0])
    prob_matrix[0,3]=count_ta/len(list_of_sequences[0])


    prob_matrix[1,0]=count_ac/len(list_of_sequences[0])
    prob_matrix[1,2]=count_gc/len(list_of_sequences[0])
    prob_matrix[1,3]=count_tc/len(list_of_sequences[0])

    prob_matrix[2,0]=count_ag/len(list_of_sequences[0])
    prob_matrix[2,1]=count_cg/len(list_of_sequences[0])
    prob_matrix[2,3]=count_tg/len(list_of_sequences[0])
    
    prob_matrix[3,0]=count_at/len(list_of_sequences[0])
    prob_matrix[3,1]=count_ct/len(list_of_sequences[0])
    prob_matrix[3,2]=count_gt/len(list_of_sequences[0])

    prob_matrix[0,0]=1-prob_matrix[1,0]-prob_matrix[2,0]-prob_matrix[3,0]
    prob_matrix[1,1]=1-prob_matrix[0,1]-prob_matrix[2,1]-prob_matrix[3,1]
    prob_matrix[2,2]=1-prob_matrix[0,2]-prob_matrix[1,2]-prob_matrix[3,2]
    prob_matrix[3,3]=1-prob_matrix[0,3]-prob_matrix[1,3]-prob_matrix[2,3]
    #acgt
    
    return prob_matrix
                       


            
                
```
